```python
import os
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor

from sklearn import metrics
import import_ipynb
import starting_out
os.chdir('C:/Users/jsbae/410project/data')
```

```python
df = starting_out.scale_data(starting_out.clean_data(pd.DataFrame(pd.read_csv('Life Expectancy Data.csv'))))
df.describe()
```

```python
y = df.Life_Expectancy
X = df.drop(columns=['Life_Expectancy'])

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8)
```

```python
model = RandomForestRegressor(n_estimators = 1000, random_state = 42)
model.fit(X_train, y_train)
```

```python
# Use the forest's predict method on the test data
predictions = model.predict(X_test)

# Print out the mean squared error (MSE )
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, predictions))
print('Mean Squared Error:', metrics.mean_squared_error(predictions, y_test))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(predictions, y_test)))
print('Coefficient of Determination R^2:', model.score(X_test, y_test))
```

```python
grad_boost = GradientBoostingRegressor()
grad_boost.fit(X_train, y_train)
```
